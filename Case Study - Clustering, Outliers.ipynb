```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans5 = KMeans(n_clusters=5)
```

```python
kmeans5
```

```python
kmeans5.fit(digits.data)
```

```python
from sklearn.metrics.cluster import completeness_score
```

```python
kmeans5.labels_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(digits.images[2])
```

```python
completeness_score(labels_true=digits.target, labels_pred=kmeans5.labels_)
```

```python
kmeans10 = KMeans(n_clusters=10)
```

```python
kmeans10.fit(digits.data)
```

```python
completeness_score(labels_true=digits.target, labels_pred=kmeans10.labels_)
```

```python
kmeans10.labels_
```

```python
import pandas as pd
```

```python
house_data = pd.read_csv('https://raw.githubusercontent.com/zekelabs/data-science-complete-tutorial/master/Data/house_rental_data.csv.txt', index_col='Unnamed: 0')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
house_data_tf = mm.fit_transform(house_data)
```

```python
kmeans3 = KMeans(n_clusters=3)
```

```python
kmeans3.fit(house_data_tf)
```

```python
house_data[kmeans3.labels_ == 2]
```

```python
from sklearn.cluster import DBSCAN,MeanShift
```

```python
dbscan = DBSCAN(eps=.4, min_samples=50)
meanshift = MeanShift()
```

```python
dbscan.fit(house_data_tf)
```

```python
meanshift.fit(house_data_tf)
```

```python
dbscan.labels_
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
ee = EllipticEnvelope(contamination=.05)
```

```python
ee.fit(house_data_tf)
```

```python
house_data[ee.predict(house_data_tf) == -1]
```
